## Cryptocurrency Scraper is a Python project focused on extracting real-time data from cryptocurrency websites. The primary goal is to provide users with up-to-date information on various cryptocurrencies, including prices, market caps, and trading volumes by employing web scraping techniques. The collected data can be subsequently utilized for further in-depth analysis.

### Import Libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/crypto'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find('tbody')
```

```python
table
```

```python
rows=table.find_all('tr')
rows
```

```python
rows=table.find_all('tr',style='')
rows
```

### Initiate storing extracted information from HTML table

```python
column_info=[]
for i in rows:
    cols=i.find_all('td')
    crypto_currency=[c.text for c in cols]
    print(crypto_currency)
    column_info.append(crypto_currency)
print(crypto_currency)
```

### Initialize empty lists to store different attributes of cryptocurrency data. Each list corresponds to a specific attribute like symbol, name, price, change, etc. For each iteration, it extracts specific elements from i and appends them to the corresponding lists.

```python
Symbol=[]
Name=[]
Price=[]
Change=[]
Percentage_Change=[]
Market_Cap=[]
Volume_in_Currency=[]
Volume_in_Currency_in_24HR=[]
Total_volume_all_Currencies=[]
Circulating_Supply=[]
for i in column_info:
    Symbol.append(i[0])
    Name.append(i[1])
    Price.append(i[2])
    Change.append(i[3])
    Percentage_Change.append(i[4])
    Market_Cap.append(i[5])
    Volume_in_Currency.append(i[6])
    Volume_in_Currency_in_24HR.append(i[7])
    Total_volume_all_Currencies.append(i[8])
    Circulating_Supply.append(i[9])
```

```python
Data={'Symbol':Symbol,'Name':Name,'Price':Price,'Change':Change,'Percentage Change':Percentage_Change,'Market Cap':Market_Cap,'Volume in Currency':Volume_in_Currency,'Volume in Currency in 24HR':Volume_in_Currency_in_24HR,'Total volume all Currencies':Total_volume_all_Currencies,'Circulating Supply':Circulating_Supply}
```

### Create a Dataframe of the table

```python
crypto_info=pd.DataFrame(Data)
crypto_info
```

### Download and store in csv format for further analysis

```python
crypto_info.to_csv('Crypto.csv',index=False)
```
